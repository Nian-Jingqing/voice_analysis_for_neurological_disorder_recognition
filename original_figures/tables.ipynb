# Calculate tables from extracted data of the literature review

```python
import os
import pandas as pd 
```

```python
# import data
all_df = pd.read_csv(os.path.join(
        'data',
        'data_extraction-v03_1-export-sorted.csv'
    ))
```

```python
# Calculate mean, median, standard deviation and minimum and maximum values
# of the patient and control groups used in the extracted studies.

# https://www.geeksforgeeks.org/pandas-groupby-one-column-and-get-mean-min-and-max-values/
df_subject_numbers = all_df.groupby('Disorder').agg(
    {
    #'Disorder': 'rank',
    'diseased_all': ['mean', 'median', 'std', 'min', 'max'], 
    'control_all': ['mean', 'median', 'std', 'min', 'max']}
)
```
